```python
import os
import PyPDF2
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from PyPDF2.errors import PdfReadError

# Make sure to download the necessary NLTK data
import nltk
nltk.download('punkt')
nltk.download('stopwords')

# Step 2: Extract text and metadata from each PDF file
data = []
for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf_file)
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            full_text = ""
            for page in reader.pages:
                text = page.extract_text()
                if text:  # Check if text extraction is successful
                    full_text += text

            # Extract metadata with error handling
            metadata = reader.metadata or {}
            author = metadata.get('/Author', 'Unknown')
            subject = metadata.get('/Subject', 'Unknown')

            data.append({'Author': author, 'Subject': subject, 'Text': full_text})
    except PdfReadError:
        print(f"Error reading {pdf_file}. The file may be corrupted.")

# Step 3: Text processing
processed_data = []
stop_words = set(stopwords.words('english'))

for item in data:
    tokens = word_tokenize(item['Text'])
    tokens = [word.lower() for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stop_words]
    processed_text = ' '.join(tokens)
    item[''] = processed_text
    processed_data.append(item)

# Step 4: Create a DataFrame
df2 = pd.DataFrame(processed_data)

print(df2)
```

```python
# Load the spaCy model for lemmatization
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])

# Step 1: Navigate to the 'GCSE' folder and retrieve the list of PDF files
pdf_folder = 'C:\\Users\\Asha\\GCSE'
pdf_files = [file for file in os.listdir(pdf_folder) if file.endswith('.pdf')]

# Step 2: Extract text and metadata from each PDF file
data = []
for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf_file)
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            full_text = ""
            for page in reader.pages:
                text = page.extract_text()
                if text:  # Check if text extraction is successful
                    full_text += text

            # Extract metadata with error handling
            metadata = reader.metadata or {}
            author = metadata.get('/Author', 'Unknown')
            subject = metadata.get('/Subject', 'Unknown')

            # Append the file name along with other metadata
            data.append({'File_Name': pdf_file, 'Author': author, 'Subject': subject, 'Text': full_text})
    except PdfReadError:
        print(f"Error reading {pdf_file}. The file may be corrupted.")

# Step 3: Text processing with lemmatization and stopwords removal
processed_data = []
stop_words = set(stopwords.words('english'))

for item in data:
    doc = nlp(item['Text'])
    tokens = [token.lemma_ for token in doc if token.pos_ in ["NOUN", "ADJ", "VERB", "ADV"]]
    tokens = [word.lower() for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stop_words]
    processed_text = ' '.join(tokens)

    # Add the processed text, author, subject, and file name to the final dataset
    processed_data.append({'File_Name': item['File_Name'], 'Author': item['Author'], 'Subject': item['Subject'], 'Text': item['Text'], 'Processed_Text': processed_text})

# Step 4: Create a DataFrame with File_Name, Author, Subject, and Processed_Text
df_lem = pd.DataFrame(processed_data)

# Print the DataFrame to check the output
print(df_lem)
```

```python
# Define a function to assign the correct author label
def assign_author(text):
    text = text.lower()  # Convert text to lowercase for case-insensitive matching
    if 'aqa' in text:
        return 'AQA'
    elif 'ocr' in text:
        return 'OCR'
    elif 'pearson' in text:  
        return 'Pearson Edexcel'
    else:
        return ''

# Apply the function to the 'File_Name' column and create or update the 'Author' column
df_lem['Author'] = df_lem['Text'].apply(assign_author)

print(df_lem)
```
